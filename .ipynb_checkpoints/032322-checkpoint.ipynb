```python
def logistic_growth(x, L, k, x0, offset):
    
    return L/(1+np.exp(-k*(x-x0))) + offset
```

```python
Exp.times
```

```python
from scipy.optimize import curve_fit

L, k, x0 = curve_fit(logistic_growth, data["time"], data["OD"], p0=(0.7, 1, 250))[0]
```

```python
plt.plot(logistic_growth(np.array(data["time"]), L, k, x0))
plt.plot(np.array(data["time"]), data["OD"])
```
